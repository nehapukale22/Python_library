```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from matplotlib import style
```

```python
#it is the data to plot bar chart
classes=["Python","R","AI","ML","DS"]
class1_students=[30,10,20,25,10]#out of 100 student in each class
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
plt.show()
```

```python
plt.barh(classes,class1_students)#barh to plot graph horizontally
plt.show()
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge")#default alignment is center
plt.show()
```

```python
#to  change color and edge color
plt.bar(classes,class1_students,width=0.2,align="edge",
        color="y",edgecolor="m")
plt.show()
```

```python
#to change edge line width of bar
plt.bar(classes,class1_students,width=0.2,align="edge",
        color="y",edgecolor="m",linewidth=5)
plt.show()
```

```python
#Matplotlib allows you to adjust the transparency of a graph plot using the alpha attribute. 
#If you want to make the graph plot more transparent, 
#then you can make alpha less than 1, such as 0.5 or 0.25.
#If you want to make the graph plot less transparent, 
#then you can make alpha greater than 1.
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",
        color="y",edgecolor="m",linewidth=5,alpha=0.6)
plt.show()
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",
        color="y",edgecolor="m",linewidth=3,alpha=0.9,linestyle="--",label="Class 1 Students")
plt.legend()
plt.show()
```

```python
#to hide the graph use visible parameter
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",
        color="y",edgecolor="m",linewidth=3,alpha=0.9,
        linestyle="--",label="Class 1 Students",visible=False)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.title("Bar Chart of IAIP Class",fontsize=30)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No. of Students",fontsize=25)

#set size of tick labels
plt.tick_params(axis="both",labelsize=20)
plt.bar(classes,class1_students,align="edge",
        color="k",edgecolor="m",linewidth=5,alpha=0.9,
        linestyle="--",label="Class 1 Students")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.title("Bar Chart of IAIP Class",fontsize=30)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No. of Students",fontsize=25)
width=0.2
#set size of tick labels
plt.tick_params(axis="both",labelsize=20)
plt.bar(classes_index,class1_students,color="k",linewidth=3,alpha=0.9,width=0.2,
        linestyle="--",label="Class 1 Students")
plt.bar(classes_index+width,class2_students,color="b",linewidth=5,alpha=0.9,width=0.2,
        linestyle="--",label="Class 2 Students")
plt.xticks(classes_index+width,classes,rotation=25)#classes_index+width replaces with classes
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.title("Bar Chart of IAIP Class",fontsize=30)
plt.xlabel("Classes",fontsize=25)
plt.ylabel("No. of Students",fontsize=25)
width=0.2
#set size of tick labels
plt.tick_params(axis="both",labelsize=20)
plt.bar(classes_index,class1_students,color="k",linewidth=3,alpha=0.9,width=0.2,
        linestyle="--",label="Class 1 Students")
plt.bar(classes_index+width,class2_students,color="b",linewidth=5,alpha=0.9,width=0.2,
        linestyle="--",label="Class 2 Students")
plt.bar(classes_index+width+width,class3_students,color="y",linewidth=5,alpha=0.9,width=0.2,
        linestyle="--",label="Class 3 Students")
plt.xticks(classes_index+width,classes,rotation=25)#classes_index+width replaces with classes
plt.legend()
plt.show()
```

```python

plt.title("Bar Chart of IAIP Class",fontsize=18)
plt.ylabel("Classes",fontsize=15)
plt.xlabel("No. of Students",fontsize=15)
width=0.2
#set size of tick labels
plt.tick_params(axis="both",labelsize=20)
plt.barh(classes_index,class1_students,color="k",alpha=0.9,label="Class 1 Students")
plt.barh(classes_index+width,class2_students,color="b",alpha=0.9,
        linestyle="--",label="Class 2 Students")
plt.barh(classes_index+width+width,class3_students,color="y",alpha=0.9,
        linestyle="--",label="Class 3 Students")
plt.yticks(classes_index+width,classes,rotation=25)#classes_index+width replaces with classes
plt.legend(loc=4)
plt.show()
```
